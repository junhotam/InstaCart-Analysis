```python
# checking for duplicates
customers.duplicated().sum()
```

```python
# checking for nan's 
customers[customers['first_name'].isnull()]

# we are missing some first names in the data, but that is not much of a worry
# and we cannot do anything about it
```

```python
# checking for mixed data types
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)

# first_name is the only col that pops up most likely because of the NaN's
```
